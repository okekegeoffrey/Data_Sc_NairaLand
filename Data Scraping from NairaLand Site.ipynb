```python
# Web scraping code
# import the necessary libraries
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as bs
from datetime import datetime
```

```python
url = 'http://www.nairaland.com/home'
scrapped_data = requests.get(url)
fetched_data = scrapped_data.content
fetched_data
```

```python
soup = bs(fetched_data, 'lxml')
soup
```

```python
# cleaner HTML page
print(soup.prettify())
```

```python
# display only the part of the data needed. Needed part are in the cells of table tags (<td>)
soup('td')
```

```python
# let us read out the texts in the tag cells in tables

for data in soup('td'):
    print(data.text)
```

```python
# since we don't need everything here, let's use the re(regular expression) module to extract relevant birthday list
# members whose birthdays are not displayed will be ignored, so that we don't have to deal with NAN values in our data

member_found = None

re_match = '[\w]+\([\d]+\)' # any word count+1 followed-by '(' followed-by any number count+1 followed-by ')'

for data in soup('td'):
    data_found = re.findall(re_match, data.text)
    
    if data_found:
        member_found = data_found

print(member_found)
```

```python
# further to separate the username from age
# use list comprehension to replace the last brace ')' with empty '' in member_found above

member_found_replace = [x.replace(")", "") for x in member_found]  # replaces ")" by ""
print(member_found_replace)
```

```python
# now split 'member_found_replace' based on '(' between the username and age
# use for-loop to loop through each item of the 'member_found_replace' list above
for y in member_found_replace:
    member_cleaned = y.split('(')
    print(member_cleaned)
    
# what we have 'member_cleaned' is individual list with two elements each
# lets combine all the lists into a dictionary
```

```python
# first declare 'member_cleaned' as empty dictionary, so we can append individual list above into it

member_cleaned = {}

for y in member_found_replace:
    temp_data = y.split('(')
    
    member_cleaned[temp_data[0]] = int(temp_data[1])
    
print(member_cleaned)
```

```python
# conv the dict 'member_cleaned' above into a pandas DataFrame
# NOTE: we have to conv the dict items into a list for it to work with Pandas DataFrame


# define the column names
column_name = ['Username', 'Age']

# conv to list and then DataFrame
df = pd.DataFrame(list(member_cleaned.items()), columns = column_name)
df
```

```python
# lets add a column for today's date
# using the datetime module

todays_date = datetime.now().date()
df['Date'] = todays_date
df
```

```python
# lets save the dataframe into csv file
df.to_csv('Nairaland_user_age.csv')
```
