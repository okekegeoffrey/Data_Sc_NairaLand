```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read-in created csv file
data = pd.read_csv('Nairaland_user_age.csv')
data
```

```python
# general information about the file
data.info()
```

```python
# visitors who are 25yrs and below
vis_25_bel = data[data['Age'] <= 25]
vis_25_bel
```

```python
# describe the imported data: mean, median, max, min
data.describe()
```

```python
# number of age-mates in different age values
data['Age'].value_counts()
```

```python
# the maximum age
data['Age'].max()
```

```python
# minimum age
data['Age'].min()
```

```python
# sort from the youngest to the oldest
data.sort_values(by = 'Age', ascending = True)
```

```python
# sort from the oldest down to the youngest
data.sort_values(by = 'Age', ascending = False) #Really? What is a 95 yr old doing on NairaLand?
```

```python
# How many years old would these users be in 10 yrs time?
data['Age_10yrs_time'] = data['Age'] + 10
data
```

```python
# area plot of their age distribution
data.plot.area()
```

```python
# remove the 'Age_10yrs_time' column
del data['Age_10yrs_time']
```
